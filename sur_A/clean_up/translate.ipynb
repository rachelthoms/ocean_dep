```python
df_ev = df_ev.head(3)

def translate (column_val):
   command = 'export OPENAI_API_KEY="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";opentranslator  --translate english --text "{value}"'
   output = sp.getoutput(command.format(value=column_val))
   return output

col_to_trans = [col for col in df_ev.columns if 'note' in col]

for col in col_to_trans:
   new_col_name = str(col)+"_trans"
   df_ev[new_col_name] = df_ev.apply(lambda row : None if pd.isnull(row[col]) else translate(row[col]), axis = 1)
   
df_ev.head()
```
