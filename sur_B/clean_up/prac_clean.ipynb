## Set up

```python
# import packages
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
from zipfile import ZipFile
import os
import seaborn as sns

pd.options.mode.chained_assignment = None
```

There is a parent table and many linked tables within the survey folder. For cleaning up the practice survey the main table (high level respondent information i.e. demographic questions) and the practice table (fishing practices used) are relavent.

```python
# unzipping survey tables

# specifying location of .zip and extraction location
sur_B_path="/home/rthoms/ocean_dep/sur_B/raw_data/Sur_B_03072023.zip"
sur_B_raw="/home/rthoms/ocean_dep/sur_B/raw_data/sur_B_raw"

# loading the .zip and creating a zip object
with ZipFile(sur_B_path, 'r') as zObject:

	# Extracting all the members of the zip
	# into a the specified location.
	zObject.extractall(
		path=sur_B_raw)
```

```python
# specifiy location of raw data
# main survey table
ss_raw = os.path.join(sur_B_raw, "Form_4_0.csv")
df_ss = pd.read_csv(ss_raw, na_values='na')

# practice table
prac_raw = os.path.join(sur_B_raw, "fish_pract_table_1.csv")
df_prac_raw = pd.read_csv(prac_raw, na_values='na')
df_ss
```

```python
# Isolate relevant columns from the main table
df_ss_sub=df_ss.loc[:, ['GlobalID','id','ccp','gender']]

# Rename GlobalID column to match with children IDs
df_ss_sub.rename(columns={'GlobalID':'ParentGlobalID'}, inplace=True)

# Join demographic table and practice table
# note that fishers who use two gears will be represented twice 
df_wk = pd.merge(df_ss_sub,df_prac_raw,on='ParentGlobalID', how='outer')
```

```python
df_wk
```

```python
# drop unnecessary columns
df_wk = df_wk[df_wk.columns[~df_wk.columns.str.startswith('_')]]
col_to_drop = ['practice', 'catch_percent','CreationDate','Creator','EditDate','Editor']
df_wk.drop(col_to_drop, axis=1, inplace=True)
df_wk
```

```python
# Examine blank entries
df_na = df_wk[df_wk.isna().any(axis=1)]
df_na
```

All rows returned are `glean` or `hand_line` practices which correctly have `NaN` for `owner` column

```python
# Examine NR (not reported)
df_nr = df_wk[df_wk.apply(lambda r: r.str.contains('NR', case=True).any(), axis=1)] 
df_nr
```

```python
# Examine catch proportion column for gleaners 
df_glean=df_wk[(df_wk['fish_pract']== 'glean') & (df_wk['catch_proportion'] != 1.00)]
df_glean
```

```python
# Examine catch proportion column for hand line
df_hl=df_wk[(df_wk['fish_pract']== 'hand_line') & (df_wk['catch_proportion'] != 1.00)]
df_hl
```

```python
# Examine catch proportion = 0
df_0=df_wk[df_wk['fish_pract']== 0]
df_wk=df_wk.drop(columns=['GlobalID','ObjectID'], axis=1)
```

```python
# categorize the practive (glean, shrimp_net, vs capture fish)
def cat_pract(pract):
    if pract == 'glean' or pract == 'shrimp_net':
        return pract
    else:
        return 'fish'
    
df_wk['pract_cat'] = df_wk.apply(lambda row : cat_pract(row['fish_pract']), axis = 1)

df_prac = df_wk.drop_duplicates(
  subset = ['id', 'pract_cat'],
  keep = 'first').reset_index(drop = True)

df_agg = df_prac.groupby(['pract_cat']).agg(sample = ('id', pd.Series.nunique)).reset_index() 
prac_dict = dict(zip(df_agg['pract_cat'], df_agg['sample']))
```
